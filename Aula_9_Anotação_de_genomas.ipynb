# Aula 9: Anotação de genomas

- **Anotação estrutural** consiste na identificação e caracterização de regiões funcionais no genoma, incluindo: genes codificantes de proteínas, genes de RNAs não codificantes, regiões de DNA repetitivo e regiões regulatórias.

- **Anotação funcional** consiste na identificação de funções biológicas e processos biológicos dos elementos funcionais de um genoma, sendo uma extensão da anotação estrutural. Isso inclui: Rotas de metabolismo primário e secundário ou ainda fatores de virulência e genes de resistência.
