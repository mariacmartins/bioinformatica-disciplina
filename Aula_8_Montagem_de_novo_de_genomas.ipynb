# Montagem *de novo* de genomas

A montagem de genomas consiste no processo de (re)construção da sequência de DNA original, tendo como objetivo a construção de *contigs*, a partir das *reads* de sequenciamento. 

Montagem *de novo*: consiste na geração de *contigs* a partir da sobreposição de *reads*, determinando a sequência de DNA quando não há sequência genômica conhecida disponível para uso (referência).

## Técnicas:
### Overlap-Layout-Consensus (OLC):
**Overlap:** Consiste na identificação de todas as possíveis sobreposições entre leituras do conjunto de dados. 

**Layout:** As leituras são organizadas e agrupadas de acordo com a sua disposição espacial identificada a partir das sobreposições. 

**Consensus:** Uma sequência consenso é produzida a partir do alinhamento múltiplo

- **Vantagens:** Preserva informação das leituras.

- **Desvantagem:** Computacionalmente custoso para casos onde há muitas leituras e leituras muito curtas (segunda geração de sequenciamento).


### Grafos de Bruijn
Neste caso, todas as subsequências com um tamanho fixo "k" presentes em um conjunto de sequências são identificadas. Estas subsequências, chamadas k-mers, são usadas como nós para construção de um grafo. 

Os nós são conectados em todos os casos onde há uma sobreposição de tamanho k-1 entre os nós.
As *contigs* são reconstruídas como caminhos neste grafo.

- **Vantagens:** Menor custo computacional.
- **Desvantagem:** Maior chance de artefatos de sequenciamento por conta dos k-mers.
