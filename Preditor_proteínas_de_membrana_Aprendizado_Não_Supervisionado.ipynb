```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('http://bit.ly/bioinfo-ml-membrane-protein-dataset')
df.head()
```

```python
X = df.drop(['membrane'], axis=1)
y = df['membrane']
```

```python
from sklearn.decomposition import PCA 

pca = PCA(n_components=2)
components = pca.fit_transform(X)

df_pca = pd.DataFrame(components, columns=['pca_component_1', 'pca_component_2'])
df_pca['membrane'] = y.astype(str)
```

```python
df_pca.head()
```

```python
fig = px.scatter(df_pca, x='pca_component_1', y='pca_component_2', color='membrane')
fig.show()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2)
components = tsne.fit_transform(X)

df_tsne = pd.DataFrame(components, columns=['tsne_components_1', 'tsne_components_2'])
df_tsne['membrane'] = y.astype(str)
df_tsne.head()
```

```python
fig = px.scatter(df_tsne, x='tsne_components_1', y='tsne_components_2', color='membrane')
fig.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import homogeneity_score, completeness_score
```

```python
kmeans = KMeans(n_clusters=10)
df_tsne['kmeans_cluster'] = kmeans.fit_predict(df_tsne[['tsne_components_1', 'tsne_components_2']]).astype(str)
df_tsne.head()
```

```python
fig = px.scatter(df_tsne, x='tsne_components_1', y='tsne_components_2', color='kmeans_cluster')
fig.show()
```

```python
homogeneity_score(y, df_tsne['kmeans_cluster'])
```

```python
completeness_score(y, df_tsne['kmeans_cluster'])
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=2)
df_tsne['dbscan_cluster'] = dbscan.fit_predict(df_tsne[['tsne_components_1', 'tsne_components_2']]).astype(str)
```

```python
fig = px.scatter(df_tsne, x='tsne_components_1', y='tsne_components_2', color='dbscan_cluster')
fig.show()
```

```python
homogeneity_score(y, df_tsne['dbscan_cluster'])
```

```python
completeness_score(y, df_tsne['kmeans_cluster'])
```
